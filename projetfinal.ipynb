```python
from IPython.display import Audio, display, clear_output, HTML
import ipywidgets as widgets

# D√©finir les clusters

def get_cluster(sleep_duration, sleep_quality, stress_percent):
    if stress_percent > 70 and sleep_duration < 6:
        cluster = "high stress & low sleep"
        recommendation = (
            "‚ö†Ô∏è Cluster: Stress √©lev√© et faible sommeil.\n"
            "Recommandations:\n"
            "‚Ä¢ Exercices de relaxation quotidiens\n"
            "‚Ä¢ R√©duire la caf√©ine apr√®s 16h\n"
            "‚Ä¢ Viser 7-8 heures de sommeil\n"
            "‚Ä¢ Activit√© physique r√©guli√®re"
        )
    elif stress_percent > 70:
        cluster = "high stress"
        recommendation = (
            "‚ö†Ô∏è Cluster: Stress √©lev√©.\n"
            "Recommandations:\n"
            "‚Ä¢ Techniques de respiration profonde\n"
            "‚Ä¢ Pauses r√©guli√®res\n"
            "‚Ä¢ Activit√© physique mod√©r√©e"
        )
    else:
        cluster = "healthy"
        recommendation = (
            "‚úÖ Cluster: Sommeil et stress corrects.\n"
            "Continuez vos bonnes habitudes!"
        )
    return cluster, recommendation

# Sons disponibles
sounds = {
    "Plage": "beach-26681.mp3",
    "Pluie": "rain-sound-188158.mp3",
    "Nature": "the-voices-of-birds-in-the-forest-7715.mp3"
}

# Fonction principale
def run_system():
    clear_output()
    print("üåô Syst√®me simple de recommandation sommeil üåô\n")

    # Widgets pour entr√©e utilisateur
    age_widget = widgets.IntSlider(
        value=30, min=10, max=100, description='√Çge:', style={'description_width': 'initial'}
    )
    sleep_duration_widget = widgets.FloatSlider(
        value=7, min=3, max=10, step=0.5, description='Dur√©e sommeil (h):', style={'description_width': 'initial'}
    )
    sleep_quality_widget = widgets.IntSlider(
        value=7, min=1, max=10, step=1, description='Qualit√© sommeil (1-10):', style={'description_width': 'initial'}
    )
    stress_widget = widgets.IntSlider(
        value=50, min=0, max=100, description='Stress %:', style={'description_width': 'initial'}
    )

    submit_btn = widgets.Button(description='Valider', button_style='success')
    input_box = widgets.VBox([age_widget, sleep_duration_widget, sleep_quality_widget, stress_widget, submit_btn])
    display(input_box)

    # Sortie principale
    output = widgets.Output()
    display(output)

    def on_submit(b):
        with output:
            clear_output()
            age = age_widget.value
            sleep_duration = sleep_duration_widget.value
            sleep_quality = sleep_quality_widget.value
            stress_percent = stress_widget.value

            cluster, rec = get_cluster(sleep_duration, sleep_quality, stress_percent)
            print(rec)

            # Si stress √©lev√©, proposer son relaxant
            if cluster in ["high stress & low sleep", "high stress"]:
                print("\nüéµ Pour vous aider √† vous d√©tendre et mieux dormir, choisissez un son relaxant ci-dessous :")

                dropdown = widgets.Dropdown(
                    options=list(sounds.keys()),
                    description='Son:',
                    style={'description_width': 'initial'}
                )
                play_btn = widgets.Button(description='‚ñ∂Ô∏è √âcouter', button_style='info')

                # Sortie sp√©cifique pour l'audio
                audio_out = widgets.Output()

                def on_play(btn):
                    with audio_out:
                        clear_output(wait=True)
                        print(f"\nüéµ Lecture du son relaxant: {dropdown.value}")
                        print("üí° Astuce: Fermez les yeux, respirez profond√©ment et √©coutez pendant 10-15 minutes.\n")
                        display(Audio(filename=sounds[dropdown.value], autoplay=True))

                play_btn.on_click(on_play)

                display(widgets.HBox([dropdown, play_btn]), audio_out)

    submit_btn.on_click(on_submit)


# Lancer le syst√®me

run_system()
```

```python
from IPython.display import display, HTML, Image, clear_output
import ipywidgets as widgets


# Fonction de diagnostic
def diagnose(bp_sys, bp_dia, hr, bmi, steps, stress_pct):
    problems = []

    if bp_sys >= 140 or bp_dia >= 90:
        problems.append({
            "title": "Hypertension art√©rielle",
            "advice": "Consultez un professionnel. R√©duire sel/caf√©ine, activit√© r√©guli√®re, suivi tension.",
            "image": "Hypertension.jpg"
        })
    elif bp_sys >= 130 or bp_dia >= 80:
        problems.append({
            "title": "Tension l√©g√®rement √©lev√©e",
            "advice": "Surveillez la tension, r√©duire stress, activit√© mod√©r√©e.",
            "image": "Hypertension.jpg"
        })

    if hr >= 100:
        problems.append({
            "title": "Tachycardie",
            "advice": "Si persistante, consulter. Repos, hydratation, √©viter stimulants.",
            "image": "tachycardie.jpg"
        })
    elif hr < 50:
        problems.append({
            "title": "Bradycardie",
            "advice": "Si sympt√¥mes ‚Üí consulter. Suivi m√©dical recommand√©.",
            "image": "bradycardie.jpg"
        })

    if bmi >= 30:
        problems.append({
            "title": "Ob√©sit√©",
            "advice": "R√©duire calories vides, augmenter activit√©, consulter nutritionniste si besoin.",
            "image": "Ob√©sit√©.jpg"
        })
    elif bmi >= 25:
        problems.append({
            "title": "Surpoids",
            "advice": "Activit√© r√©guli√®re, alimentation √©quilibr√©e, objectif perte de poids mod√©r√©e.",
            "image": "Activit√© mod√©r√©e.jpg"
        })

    if steps < 4000:
        problems.append({
            "title": "Faible activit√© physique",
            "advice": "Essayez 20‚Äì30 min de marche/jour.",
            "image": "Activit√© mod√©r√©e.jpg"
        })
    elif steps < 7000:
        problems.append({
            "title": "Activit√© mod√©r√©e",
            "advice": "Visez progressivement 7000‚Äì10000 pas/jour.",
            "image": "Activit√© mod√©r√©e.jpg"
        })

    if stress_pct >= 50:
        problems.append({
            "title": "Stress √©lev√©",
            "advice": "Respiration, pauses r√©guli√®res, sommeil r√©gulier.",
            "image": "√âtat global sain.jpg"
        })

    if not problems:
        problems.append({
            "title": "Profil sain",
            "advice": "Continuez vos bonnes habitudes.",
            "image": "√âtat global sain.jpg"
        })

    return problems


# Interface utilisateur
def build_ui():
    clear_output()
    print("üåø Syst√®me Sant√© ‚Äî Diagnostic visuel simple\n")

    # Widgets d'entr√©e avec description large
    style = {'description_width': '200px'}
    layout = widgets.Layout(width='500px')

    age = widgets.IntSlider(value=35, min=10, max=90, description='√Çge de l‚Äôutilisateur:', style=style, layout=layout)
    bp_sys = widgets.IntText(value=120, description='BP systolique (mmHg):', style=style, layout=layout)
    bp_dia = widgets.IntText(value=80, description='BP diastolique (mmHg):', style=style, layout=layout)
    hr = widgets.IntSlider(value=72, min=30, max=160, description='Fr√©quence cardiaque (bpm):', style=style, layout=layout)
    bmi = widgets.FloatSlider(value=24.0, min=12.0, max=45.0, step=0.1, description='BMI:', style=style, layout=layout)
    steps = widgets.IntText(value=5000, description='Nombre de pas/jour:', style=style, layout=layout)
    stress = widgets.IntSlider(value=40, min=0, max=100, description='Stress (%):', style=style, layout=layout)

    submit = widgets.Button(description="Analyser", button_style='success', layout=widgets.Layout(width='200px'))
    display(widgets.VBox([age, bp_sys, bp_dia, hr, bmi, steps, stress, submit]))

    out = widgets.Output()
    display(out)

    def on_submit(b):
        with out:
            clear_output()
            problems = diagnose(bp_sys.value, bp_dia.value, hr.value, bmi.value, steps.value, stress.value)

            # Afficher titre principal
            display(HTML(f"<h3>üîé R√©sultats pour l'utilisateur ‚Äî √¢ge {age.value} ans</h3>"))

            # Afficher titres d√©tect√©s
            display(HTML("<ul>" + "".join([f"<li><b>{p['title']}</b></li>" for p in problems]) + "</ul>"))

            # Afficher cartes (image + titre + conseil)
            for p in problems:
                display(Image(filename=p["image"], width=420, height=260))
                display(HTML(f"<h4>{p['title']}</h4><p>{p['advice']}</p>"))

    submit.on_click(on_submit)

# Lancer l'interface
build_ui()
```
